```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = 2
N = 5
d = p**N
```

```python
mubs  = np.load('MUBS/2-5-desarguesian.npy')
mubsK = np.load('MUBS/2-5-kantor.npy')
```

```python
mubs.shape
```

```python
for i in range(d+1):
    for j in range(d+1):
        inps = np.square(np.abs(
            mubs[i*d:(i+1)*d].conj().T @ mubsK[j*d:(j+1)*d]
        ))
        print(i, j, np.unique(np.round(inps, 5)))
```
